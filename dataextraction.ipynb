```python
from bs4 import BeautifulSoup
import os
import pandas as pd
```

```python
columns = ['company_name', 'job_role', 'job_location', 'salary', 'skills','company_rating']
df = pd.DataFrame(columns=columns)
```

```python
try:
    with open('./data/0.html', encoding='utf-8') as f:
        html_doc = f.read()
    soup = BeautifulSoup(html_doc, 'html.parser')

    
    job_cards = soup.find_all('div', attrs={'class': 'JobCard_jobCardContainer__arQlW'})
    
    for job_card in job_cards:
        
        company_names = job_card.find('span', attrs={'class': 'EmployerProfile_compactEmployerName__9MGcV'})
        job_role = job_card.find('a', attrs={'class': 'JobCard_jobTitle__GLyJ1'})
        job_location = job_card.find('div', attrs={'class': 'JobCard_location__Ds1fM'})
        salary = job_card.find('div', attrs={'class': 'JobCard_salaryEstimate__QpbTW'})
        skills_div = job_card.find('div', string=lambda x: x and 'Skills' in x)
        rating= job_card.find('span', attrs={'class': 'rating-single-star_RatingText__XENmU'})  
        
        
        
        company_name_text = company_names.text.strip() if company_names else None
        job_role_text = job_role.text.strip() if job_role else None
        job_location_text = job_location.text.strip() if job_location else None
        salary_text = salary.text.strip() if salary else None
        skills_text = skills_div.text.replace('Skills', '').strip() if skills_div else None
        rating_text=rating.text if rating else None
        
        
        job_data = {
            'company_name': company_name_text,
            'job_role': job_role_text,
            'job_location': job_location_text,
            'salary': salary_text,
            'skills': skills_text,
            'company_rating':rating_text
        }
        
        
        df.loc[len(df)] = job_data

except Exception as e:
    print(e)
```
