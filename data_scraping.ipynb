```python
import undetected_chromedriver as uc 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os

```

```python
driver = uc.Chrome() 
queries = [
    "software-engineer",
    "cloud-engineer",
    "devops-engineer",
    "database-administrator",
    "full-stack-developer",
    "backend-developer",
    "frontend-developer",
    "mobile-app-developer",
    "game-developer",
    "machine-learning-engineer",
    "network-engineer",
    "it-support-specialist",
    "security-engineer",
    "blockchain-developer",
    "embedded-systems-engineer",
    "robotics-engineer",
    "business-intelligence-analyst",
    "ui-ux-designer",
    "digital-marketing-specialist",
    "data-analyst",
    "product-manager",
    "technical-writer",
    "site-reliability-engineer",
    "artificial-intelligence-researcher",
    "penetration-tester"
]
 
output_dir = './data'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python


for query in queries:
    
    url = f'https://www.glassdoor.com/Job/india-{query}-jobs-SRCH_IL.0,5_IN115_KO6,{6+len(query)}.htm'
    driver.get(url)  

    try:
        while True:
            try:

                show_more_button = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "//button[contains(@class, 'button_Button__MlD2g') and contains(@class, 'button-base_Button__knLaX') and contains(., 'Show more jobs')]"))
                )


                actions = ActionChains(driver)
                actions.move_to_element(show_more_button).perform()


                show_more_button.click()
                time.sleep(2)  

            except Exception as e:
                print("Show more button not found, trying to close any pop-ups.")


                try:
                    close_button = WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.CLASS_NAME, "CloseButton"))
                    )


                    actions.move_to_element(close_button).perform()


                    close_button.click()
                    time.sleep(2)  

                except Exception as e:
                    print("No close button found. Exiting loop.")
                    break  

    except Exception as e:
        print(f"An error occurred: {e}")
    
    elems = driver.find_elements(By.CLASS_NAME, 'JobCard_jobCardContainer__arQlW')
    print(f'{len(elems)} items found for {query} jobs')

    with open(f'./data/{query}.html', 'w', encoding = 'utf-8') as f:
        for elem in elems:
            d = elem.get_attribute('outerHTML')

            f.write(d)
    

       



```

### We did the above process iteratively to get the maximum items. We got 900 items in every jobs except for some jobs like penetration testing and artificial intelligence researcher jobs.
